```
import numpy as np
import matplotlib.pyplot as plt

data    = np.genfromtxt("data.txt", delimiter=',')

x       = data[:, 0]
y       = data[:, 1]
label   = data[:, 2]

x_label0    = x[label == 0]
x_label1    = x[label == 1]

y_label0    = y[label == 0]
y_label1    = y[label == 1]

#Plot the training data
plt.figure(figsize=(6, 6))
plt.scatter(x_label0, y_label0, alpha=0.3, c='b')
plt.scatter(x_label1, y_label1, alpha=0.3, c='r')
plt.show()
```

```
#initialize step_size and theta0,1,2

step_size = 0.000925
theta0 = 0
theta1 = 10
theta2 = 15
```

```
#define logistic regression functions
def z_function(theta0, theta1, theta2, x, y):
  return theta0 + theta1 * x + theta2 * y

def sigmoid(theta0, theta1, theta2, x, y):
  z = z_function(theta0, theta1, theta2, x, y)
  return 1/(1+np.exp(-z))
```

```
#object function

def object_function(m, theta0, theta1, theta2):
  return (1/m) *  sum((-label[i] * np.log(sigmoid(theta0, theta1, theta2, x[i], y[i]))) - ((1-label[i]) * (np.log(1-sigmoid(theta0, theta1, theta2, x[i], y[i]))))  for i in range(m))
```

```
e_values = []
theta0_values = []
theta1_values = []
theta2_values = []
```

```
#Loop while the object function convergence 

while(1):
  j = object_function(100, theta0, theta1, theta2)
  
  new_theta0 = theta0 - step_size * (1/100) * sum((sigmoid(theta0, theta1, theta2, x[i], y[i]) - label[i]) for i in range(100))
  new_theta1 = theta1 - step_size * (1/100) * sum(((sigmoid(theta0, theta1, theta2, x[i], y[i]) - label[i]) * x[i]) for i in range(100))
  new_theta2 = theta2 - step_size * (1/100) * sum(((sigmoid(theta0, theta1, theta2, x[i], y[i]) - label[i]) * y[i]) for i in range(100))
  
  new_j = object_function(100, new_theta0, new_theta1, new_theta2)

  e_values.append(new_j)
  theta0_values.append(new_theta0)
  theta1_values.append(new_theta1)
  theta2_values.append(new_theta2)

  theta0 = new_theta0
  theta1 = new_theta1
  theta2 = new_theta2

  #until convergence
  print(new_j)
  if abs(j-new_j) < 0.000001:
    break
```

```
#plotting model parameters

theta0_values = np.array(theta0_values)
theta1_values = np.array(theta1_values)
theta2_values = np.array(theta2_values)

plt.title('Model Parameter')
plt.plot(theta0_values, 'red', label = theta0)
plt.plot(theta1_values, 'green', label = theta1)
plt.plot(theta2_values, 'blue',  label = theta2)
plt.show()

#Plot the parameters with range(0,10000), "more detailed view!!"
plt.title('Model Parameter')
plt.xlim(0,10000)
plt.plot(theta0_values, 'red', label = theta0)
plt.plot(theta1_values, 'green', label = theta1)
plt.plot(theta2_values, 'blue',  label = theta2)
plt.show()
```

```
#plotting training errors

e_values = np.array(e_values)
plt.title('Training Errors')
plt.plot(e_values,'b')
plt.xlim(0,100000)
plt.show()
```
